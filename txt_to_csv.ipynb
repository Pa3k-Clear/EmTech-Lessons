```python
f = open("/content/sample_data/naia-observatory-example.txt", "r")
eather_tet = f.read()
f.close()
print(eather_tet)
```

```python
import re
```

```python
day_reading = re.findall("[>>].\s[2].*\n", eather_tet)
day_reading
```

```python
day0 = day_reading[0]
day0
```

```python
eather_dict = {
    'date': [],
    'rainfall': [],
    'temp_max': [],
    'temp_min': [],
    'temp_mean': [],
    'wind_speed': [],
    'wind_dir': [],
}
```

```python
day0.split()
```

```python
word_convert ={
    'RAINFALL':'rainfall',
    'TMAX':'temp_max',
    'TMIN':'temp_min',
    'TMEAN':'temp_mean',
    'WNDSPD':'wind_speed',
    'WNDDR':'wind_dir'
}
```

```python
day0.split()[2:]
```

```python
dummy_eather_dict = eather_dict.copy()

tet_day = day0.split()

dummy_eather_dict['date'].append(tet_day[1])

for data in tet_day[2:]:
  subdata = data.split(':')
  key = subdata[0]
  val = float(subdata[1])

  dummy_eather_dict[word_convert[key]].append(val)

dummy_eather_dict
```

```python
eather_data = eather_dict.copy()
for reading in day_reading:
  day_data = reading.split()

  eather_data['date'].append(day_data[1].strip())
  for data in day_data[2:]:
    subdata = data.split(':')
    key, val = subdata[0], subdata[1].strip()
    eather_data[word_convert[key]].append(val)
  eather_data
```

```python
eather_data.keys()
```

```python
import pandas as pd

eather_df = pd.DataFrame(eather_data)
eather_df.head(5)
```

```python
eather_df.dtypes
```

```python
eather_df = eather_df.replace('T', -1)
eather_df.head(5)
```

```python
import numpy as np
eather_df = eather_df.replace('nan', np.NaN)
eather_df.head(5)
```

```python
eather_df.isna()
```

```python
eather_df.isna().sum()
```

```python
eather_df['date'] = pd.to_datetime(eather_df['date'])
eather_df = eather_df.set_index('date', drop=True)
eather_df
```

```python
eather_df = eather_df.apply(pd.to_numeric)
eather_df.head(5)
```

```python
eather_df.dtypes
```
