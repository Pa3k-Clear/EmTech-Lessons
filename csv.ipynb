```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

```python
eather_df = pd.read_csv('/content/clean_eather_data.csv', index_col='date')
eather_df.head(5)
```

```python
eather_df.plot()
```

```python
plt.figure(figsize=(10,5))
plt.plot(eather_df.index, eather_df['rainfall'])
plt.show
```

```python
eather_df = eather_df.set_index(pd.to_datetime(eather_df.index))
```

```python
plt.figure(figsize=(10,5))
plt.plot(eather_df.index, eather_df['rainfall'])
plt.show()
```

```python
plt.figure(figsize=(20,5))

plt.plot(eather_df.index, eather_df['temp_max'], label='Max Temp.')
plt.plot(eather_df.index, eather_df['temp_min'], label='Min Temp.')
plt.plot(eather_df.index, eather_df['temp_mean'], label='Mean Temp.')

plt.title('Temperature in 2017')

plt.xlabel('Dates (2017)')
plt.ylabel('Temperature (C)')

plt.grid()
plt.legend()

plt.show()
```

```python
eather_df['temp_max'] = eather_df['temp_max'].interpolate()
eather_df['temp_min'] = eather_df['temp_min'].interpolate()
eather_df['temp_mean'] = eather_df['temp_mean'].interpolate()
```

```python
plt.figure(figsize=(20,5))

plt.plot(eather_df.index, eather_df['temp_max'], label='Max Temp.')
plt.plot(eather_df.index, eather_df['temp_min'], label='Min Temp.')
plt.plot(eather_df.index, eather_df['temp_mean'], label='Mean Temp.')

plt.title('Temperature in 2017 (Imputed NaNs with Interpolated Values)')

plt.xlabel('Dates (2017)')
plt.ylabel('Temperature (C)')

plt.grid()
plt.legend()

plt.show()
```
